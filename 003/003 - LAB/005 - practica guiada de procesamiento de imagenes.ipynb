```python
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage as ski


def mostrar_imagenes(imagenes, titulos, cmap=None, figsize=(14, 4)):
    fig, axes = plt.subplots(1, len(imagenes), figsize=figsize)
    if len(imagenes) == 1:
        axes = [axes]

    for ax, imagen, titulo in zip(axes, imagenes, titulos):
        if imagen.ndim == 2:
            ax.imshow(imagen, cmap=cmap or "gray")
        else:
            ax.imshow(imagen)
        ax.set_title(titulo)
        ax.axis("off")

    plt.tight_layout()
    plt.show()


def segmentar_por_umbral(imagen, umbral):
    mascara = np.zeros_like(imagen, dtype=np.uint8)
    mascara[imagen > umbral] = 255
    return mascara
```

```python
base = Path(".")

cv2.imwrite(str(base / "paisaje.png"), cv2.cvtColor(ski.data.astronaut(), cv2.COLOR_RGB2BGR))
cv2.imwrite(str(base / "texto.png"), ski.data.page())
cv2.imwrite(str(base / "monedas.png"), ski.data.coins())

print("Archivos de práctica listos: paisaje.png, texto.png y monedas.png")
```

## Ejercicio 3: Bordes y umbralización

1. Cargá `texto.png` en escala de grises.
2. Aplicá `Canny` con dos pares de umbrales distintos.
3. Cargá `monedas.png` en escala de grises.
4. Segmentá la imagen de monedas con al menos dos umbrales.
5. Compará los resultados y explicá cuál te parece más útil.


```python
img_texto = cv2.imread("texto.png", cv2.IMREAD_GRAYSCALE)

canny_a = cv2.Canny(img_texto, 50, 150)
canny_b = cv2.Canny(img_texto, 100, 200)

mostrar_imagenes(
    [img_texto, canny_a, canny_b],
    ["Texto original", "Canny A", "Canny B"],
    figsize=(15, 4),
)

img_monedas = cv2.imread("monedas.png", cv2.IMREAD_GRAYSCALE)

seg_a = segmentar_por_umbral(img_monedas, 50)
seg_b = segmentar_por_umbral(img_monedas, 100)

mostrar_imagenes(
    [img_monedas, seg_a, seg_b],
    ["Monedas originales", "Umbral A", "Umbral B"],
    figsize=(15, 4),
)
```

Respuestas:

¿Cuándo conviene mirar canales por separado?
Cuando necesitás aislar información de color específica, por ejemplo para detectar objetos por su tono (como el traje naranja del ejercicio 1). Si analizás la imagen completa en BGR/RGB, los tres canales se mezclan y es más difícil segmentar por color. Separar canales también es útil antes de aplicar filtros o umbrales sobre una componente concreta.

¿Qué cambia al modificar los umbrales de Canny?
Canny usa dos umbrales (mínimo y máximo) para decidir qué gradientes de intensidad se consideran bordes. Con umbrales bajos (Canny A en tu imagen) se detectan más bordes, incluyendo ruido y detalles finos → resultado más "ruidoso". Con umbrales altos (Canny B) solo se retienen los bordes más marcados → resultado más limpio pero puede perder detalles. El tradeoff es siempre sensibilidad vs. ruido.

¿Por qué la segmentación por umbral funciona bien en una imagen y mal en otra?
Porque depende del contraste y la iluminación uniforme. En las monedas (Umbral B) funcionó bien porque hay un fondo oscuro homogéneo y las monedas son más claras → el umbral separa limpiamente los dos grupos de píxeles. En cambio, si la iluminación es despareja (sombras, gradientes de luz) o el objeto y el fondo tienen intensidades similares, el mismo umbral va a clasificar mal muchos píxeles. Es lo que se ve en Umbral A: el fondo no es uniforme y "contamina" la segmentación.
